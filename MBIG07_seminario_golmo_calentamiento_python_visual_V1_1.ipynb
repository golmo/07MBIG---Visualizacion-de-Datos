```python


data['Annual_Mean']=data['Annual_Mean'].fillna(0.0)
data['5-year_Mean']=data['5-year_Mean'].fillna(0.0)

#data1

data1['metANN']=data1['metANN'].fillna(0.0)
data1['D-J-F']=data1['D-J-F'].fillna(0.0)
data1['J-J-A']=data1['J-J-A'].fillna(0.0)

data1  = pd.concat( [data1['YEAR'],data1['D-J-F'],data1['J-J-A'],data1['metANN']],axis=1)\
[(data1['D-J-F'].abs() < 50) & (data1['J-J-A'].abs() < 50)&( data1['metANN'].abs() < 50)]

#data2

data2['metANN']=data2['metANN'].fillna(0.0)
data2['D-J-F']=data2['D-J-F'].fillna(0.0)
data2['J-J-A']=data2['J-J-A'].fillna(0.0)

data2  = pd.concat( [data2['YEAR'],data2['D-J-F'],data2['J-J-A'],data2['metANN']],axis=1)\
[(data2['D-J-F'].abs() < 50) & (data2['J-J-A'].abs() < 50)&( data2['metANN'].abs() < 50)]

#data3
data3['metANN']=data3['metANN'].fillna(0.0)
data3['D-J-F']=data3['D-J-F'].fillna(0.0)
data3['J-J-A']=data3['J-J-A'].fillna(0.0)

data3  = pd.concat( [data3['YEAR'],data3['D-J-F'],data3['J-J-A'],data3['metANN']],axis=1)\
[(data3['D-J-F'].abs() < 50) & (data3['J-J-A'].abs() < 50)&( data3['metANN'].abs() < 50)]

#data4
data4['metANN']=data4['metANN'].fillna(0.0)
data4['D-J-F']=data4['D-J-F'].fillna(0.0)
data4['J-J-A']=data4['J-J-A'].fillna(0.0)

data4  = pd.concat( [data4['YEAR'],data4['D-J-F'],data4['J-J-A'],data4['metANN']],axis=1)\
[(data4['D-J-F'].abs() < 50) & (data4['J-J-A'].abs() < 50)&( data4['metANN'].abs() < 50)]

#data5
data5  = pd.concat( [data5['Year'],round(data5['CSIRO Adjusted Sea Level']*2.54,3),\
                     round(data5['Lower Error Bound']*2.54,3),\
                     round(data5['Upper Error Bound']*2.54,3)],axis=1)

#data6
# http://wgms.ch/data_databaseversions/
#Average cumulative mass balance of “reference” Glaciers worldwide from 1945-2014 sourced from US EPA and 
#the World Glacier Monitoring Service (WGMS). This is cumulative change in mass balance of a set of “reference”
#glaciers worldwide beginning in 1945. The values represents the average of all the glaciers that were measured. 
#Negative values indicate a net loss of ice and snow compared with the base year of 1945. 
#For consistency, measurements are in meters of water equivalent, which represent changes in the average 
#thickness of a glacier.
#Year Mean cumulative mass balance Number of observations
data6['Mean cumulative mass balance']=data6['Mean cumulative mass balance'].fillna(0.0)
data6['Number of observations']=data6['Number of observations'].fillna(0.0)


#data7
data7 = pd.concat( [ data7['Date'], data7['Average'], data7['Interpolated'], data7['Trend']],axis=1)\
 [ ((data7['Date'].str.contains('-01-01')) | (data7['Date'].str.contains('-06-01'))) & (data7['Average'] > 0 )]

#data8
data8 = pd.concat([ data8['Year'], data8['Land'],data8['Land and Ocean'], data8['N Hem'], data8['S Hem']],axis=1)

#data9
#Solo valores consolidados a nivel global, ya en el dataset

data9d = pd.concat( [ data9['Year'], data9['Value'] ],axis=1)[ data9['Country Name'] == 'World']

#data10
#Ejemplo de "groupby" : sumarizar columnas numericas por 'Year' para todos los paises
#data10g  = data10[ ['Year','Total','Solid Fuel','Liquid Fuel','Gas Fuel','Cement','Gas Flaring']]\
#                 .groupby('Year')
#data10gd = data10g.apply(pd.DataFrame) 
#data10gd
#data10gd[(data10gd['Year'] ==1900)]

data10g1 = data10.groupby(['Year']).sum().reset_index()
#type(data10gd) -> es un DataFrame 
data10g2  = pd.concat([ data10g1['Year'],data10g1['Total'],data10g1['Solid Fuel'],data10g1['Liquid Fuel'],\
                        data10g1['Gas Fuel'],data10g1['Cement'],data10g1['Gas Flaring'] ],axis=1)
```

```python
### Plotting
sns.set(font_scale=0.8)

data10gf=data10g2.melt('Year',var_name='CO2 globales fuentes emisoras',value_name='Millones Tm3')
g10=sns.factorplot(x='Year',y='Millones Tm3',hue='CO2 globales fuentes emisoras',\
                   data=data10gf,aspect=4, kind='point')
g10.set_xticklabels(step=10)

sns.set(font_scale=1.1)
data9f=data9d.melt('Year',var_name='Gases invernadero globales(Equiv CO2)',\
                         value_name='Millones Tm3')
g9=sns.factorplot(x='Year',y='Millones Tm3',hue='Gases invernadero globales(Equiv CO2)',\
                  data=data9f,aspect=4,kind='point')
g9.set_xticklabels(step=2)

sns.set(font_scale=1.0)
data8f=data8.melt('Year',var_name='Equilibrio temp Tierra-Mar',value_name='Equilibrio temp Tierra-Mar valores')
g8=sns.factorplot(x='Year',y='Equilibrio temp Tierra-Mar valores',hue='Equilibrio temp Tierra-Mar',\
                  data=data8f,aspect=4,kind='point')
g8.set_xticklabels(step=10)

sns.set(font_scale=1.0)
data7f=data7.melt('Date',var_name='CO2 ppm metrics',value_name='CO2 ppm metrics values')
g7=sns.factorplot(x='Date',y='CO2 ppm metrics values',hue='CO2 ppm metrics',\
                  data=data7f,aspect=4,kind='bar')
g7.set_xticklabels(step=7)

sns.set(font_scale=1.2)
data6f=data6.melt('Year',var_name='WGMS Glaciar_varpct_1945',value_name='WGMS Glaciar_varpct_1945_var')
g6=sns.factorplot(x='Year',y='WGMS Glaciar_varpct_1945_var',hue='WGMS Glaciar_varpct_1945',\
                  data=data6f,aspect=3.5,kind='bar')
g6.set_xticklabels(step=3)

sns.set(font_scale=1.2)
data5f=data5.melt('Year',var_name='CSIRO Adjusted Sea Level_Meds',value_name='CSIRO Adjusted Sea Level_cm')
g5=sns.factorplot(x='Year',y='CSIRO Adjusted Sea Level_cm',hue='CSIRO Adjusted Sea Level_Meds',\
                  data=data5f,aspect=3.5,kind='point')
g5.set_xticklabels(step=10)

sns.set(font_scale=1.2)
data4f=data4.melt('YEAR',var_name='BaseFaraday_antartida_T_Meds',value_name='BaseFaraday_antartida_T_Meds_val')
g4=sns.factorplot(x='YEAR',y='BaseFaraday_antartida_T_Meds_val',hue='BaseFaraday_antartida_T_Meds',data=data4f,aspect=3.5,kind='bar')
g4.set_xticklabels(step=5)

sns.set(font_scale=1.2)
data3f=data3.melt('YEAR',var_name='Antartico_Atlantico_OrcadasSUR_T_Meds',value_name='Antartico_Atlantico_OrcadasSUR_T_Meds_val')
g3=sns.factorplot(x='YEAR',y='Antartico_Atlantico_OrcadasSUR_T_Meds_val',hue='Antartico_Atlantico_OrcadasSUR_T_Meds',data=data3f,aspect=3.5,kind='bar')
g3.set_xticklabels(step=5)

sns.set(font_scale=1.2)
data2f=data2.melt('YEAR',var_name='Artico_VardoNORW_T_Meds',value_name='Artico_VardoNORW_T_Meds_val')
g2=sns.factorplot(x='YEAR',y='Artico_VardoNORW_T_Meds_val',hue='Artico_VardoNORW_T_Meds',data=data2f,aspect=3.5,kind='bar')
g2.set_xticklabels(step=10)

sns.set(font_scale=1.2)
data1f=data1.melt('YEAR',var_name='Artico_Karmaku_T_Meds',value_name='Artico_Karmaku_T_Meds_val')
g1=sns.factorplot(x='YEAR',y='Artico_Karmaku_T_Meds_val',hue='Artico_Karmaku_T_Meds',data=data1f,aspect=3.5,kind='bar')
g1.set_xticklabels(step=5)

#g.map(plt.axhline,y=0.0,ls=".")
sns.set(font_scale=1.2)
dataf=data.melt('Year',var_name='MAR_T_Meds',value_name='MAR_T_Meds_val')
g=sns.factorplot(x='Year',y='MAR_T_Meds_val',hue='MAR_T_Meds',data=dataf,aspect=3.5,kind='bar')
g.set_xticklabels(step=10)
```
